```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data
df = pd.read_csv(r'C:\Users\atkhg\Downloads\movies.csv')
```

```python
# let's look at the data

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Dropping missing data for demonstration purposes

df = df.dropna()
```

```python
# Removing duplicates

df.drop_duplicates()
```

```python
# Creating the year correct column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Ordering the data by gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Predicting budget has a high correlation to gross

# Predicting company has a high correlation to gross
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Changing non-numerical columns to numerical values

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# Correlation for df_numerized

df_numerized.corr(numeric_only = 'false')
```

```python
# Correlation Matrix for df_numerized

correlation_matrix = df_numerized.corr(numeric_only = 'false')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
# Company has low correlation with gross
```

```python
correlation_mat = df_numerized.corr(numeric_only = 'false')

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```
